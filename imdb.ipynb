### 8. Какие актеры чаще всего снимаются в одном фильме вместе

```python
actors_list = array([x.split('|') for x in df.cast]).flat

pairs_set = set()

for i in actors_list:
    for n in i:
        for y in i:
            if y == n:
                continue
            else:
                if ((n,y) in pairs_set) or ((y,n) in pairs_set):
                    continue
                else:
                    pairs_set.add((n,y))
                    
pairs_dict = {x:0 for x in pairs_set}

for i in pairs_dict:
    for n in df.cast:
        if (i[0] in n.split('|')) and (i[1] in n.split('|')):
            pairs_dict[i] += 1
```

```python
lst = []
for i, v in pairs_dict.items():
    if v == max(pairs_dict.values()):
        lst.append(i)
lst
```

```python
pd.Series(pairs_dict).sort_values(ascending = False).head(5)
```

### 9. Какая студия заработала больше всего денег в жанре комедий за все время?

```python
comedy_df = df.copy()

def comedy_or_not(x):
    if 'Comedy' in x.split('|'):
        return 1
    else: 
        return 0

comedy_df['genres'] = comedy_df['genres'].apply(comedy_or_not)
comedy_df.rename(columns = {'genres':'comedy'}, inplace = True) 
comedy_df = comedy_df[comedy_df.comedy > 0]
comedy_df['profit'] = comedy_df.revenue - comedy_df.budget

revenue_dict = {}

for companies, profit in zip(comedy_df.production_companies, comedy_df.profit):
    for comp in companies.split('|'):
        try:
            revenue_dict[comp] += round(profit/len(companies.split('|')))
        except:
            revenue_dict[comp] = round(profit/len(companies.split('|')))
```

```python
pd.Series(revenue_dict).sort_values(ascending = False).head(5)
```
