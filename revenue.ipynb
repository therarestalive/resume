```python
devices = []

for i in metric_data:
    devices_dict = {
        'deviceCategory':i['dimensions'][1]['name'],
        'mobilePhone':i['dimensions'][2]['name'],
        'mobilePhoneModel':i['dimensions'][3]['name'],
        'users':i['metrics'][0]
    }
    devices.append(devices_dict)
    
metric_df = pd.DataFrame(devices)

metric_df.mobilePhone = metric_df.mobilePhone.apply(lambda x: 'Unknown' if x is None else x)
metric_df.mobilePhoneModel = metric_df.mobilePhoneModel.apply(lambda x: 'Unknown' if x is None else x)

metric_df
```

```python
test_metric_df = pd.DataFrame(metric_df.groupby(['deviceCategory', 'mobilePhone', 'mobilePhoneModel'],as_index=False)['users'].sum())
test_metric_df = test_metric_df[test_df.users > 2]
```
